```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
file_name = '/content/drive/MyDrive/ML_mastery/dataset/train.csv'
dataframe = pd.read_csv(file_name)
```

```python
dataframe
```

결측치 확인

```python
def check_missing_col(dataframe):
    missing_col = []
    counted_missing_col = 0
    for i, col in enumerate(dataframe.columns):
        missing_values = sum(dataframe[col].isna())
        is_missing = True if missing_values >= 1 else False
        if is_missing:
            counted_missing_col += 1
            print(f'결측치가 있는 컬럼은: {col}입니다')
            print(f'해당 컬럼에 총 {missing_values}개의 결측치가 존재합니다.')
            missing_col.append([col, dataframe[col].dtype])
    if counted_missing_col == 0:
        print('결측치가 존재하지 않습니다')
    return missing_col

missing_col = check_missing_col(dataframe)
```

결측치가 존재하는 행을 과감하게 없앱니다.

```python
dataframe = dataframe.dropna(axis=0)
```

범주형 데이터들의 종류가 많지 않기 때문에 원 핫 인코딩했습니다.

```python
dataframe = pd.get_dummies(dataframe)
```

```python
dataframe
```

column id는 학습에 필요없고, y= body mass로 설정

```python
import numpy as np
array = dataframe.values
X = np.concatenate([array[:,1:6],array[:,7:]],axis=1)
y = array[:,6]
```

```python
X.shape, y.shape
```

# spot-check


### linear model

```python
kfold = KFold(n_splits =10,random_state=7,shuffle=True)
scoring = 'neg_mean_squared_error'
scores =list()
models =list()
```

LR

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
model = Ridge()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

LASSO

```python
from sklearn.linear_model import Lasso
model = Lasso()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

ElasticNet

```python
from sklearn.linear_model import ElasticNet
model = ElasticNet()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

### non-linear model

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

CART

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```

SVM

```python
from sklearn.svm import SVR
model = SVR()
results = cross_val_score(model,X,y,cv=kfold,scoring = scoring)
scores.append(results)
models.append(model)
print(results.mean())
```
